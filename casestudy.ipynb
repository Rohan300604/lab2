```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='median')
null_columns=df.select_dtypes(include='float')
imputer.fit(null_columns)
null_columns=imputer.transform(null_columns)
null_columns
```

```python
null_columns=pd.DataFrame(null_columns,columns=['No of Bathrooms','Latitude','No of Bedrooms','No of Floors','Sale Price'])
null_columns
```

```python
q1=np.percentile(null_columns['Sale Price'],25,method='midpoint')
q2=np.percentile(null_columns['Sale Price'],50,method='midpoint')
q3=np.percentile(null_columns['Sale Price'],75,method='midpoint')
print(q1,q2,q3)
```

```python
iqr=q3-q1
lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr
print(lower_limit,upper_limit)
```

```python
outliers=[]
for row in df['Sale Price']:
  if row<lower_limit or row>upper_limit:
    outliers.append(row)
print(outliers)
```

```python
ind1=df[df['Sale Price']<lower_limit].index
iind2=df[df['Sale Price']>upper_limit].index
print(ind1,iind2)
```
