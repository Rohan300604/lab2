**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

````python
import pandas as pd
import numpy as np
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
print(df)
````

Histogram

````python
plt.hist(df['Glucose'], bins=10, color='blue', edgecolor='black')
plt.xlabel('Glucose')
plt.ylabel('level')
plt.title('Histogram of Glucose')
plt.show()
````

This indicates the most common glucose level range. A wider spread suggests a larger variation in glucose levels.

````python
plt.hist(df['BloodPressure'], bins=10, color='red', edgecolor='black')
plt.xlabel('BloodPressure	')
plt.ylabel('range')
plt.title('Histogram of BloodPressure')
plt.show()
````

The histogram of BloodPressure displays the distribution of blood pressure measurements across different ranges. With 10 bins, the red bars highlight the frequency of occurrences in each range, while the black edges delineate the boundaries of these bins. The distribution of blood pressure values can provide insights into whether the data is skewed towards higher or lower values, and help identify any potential outliers or patterns within the dataset.
